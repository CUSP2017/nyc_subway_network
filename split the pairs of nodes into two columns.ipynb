```python
import pandas as pd
```

```python
delay= pd.read_csv('delay_two_nodes_removal.csv',index_col=0)
```

```python
len(delay)
```

```python
master=pd.read_csv('network_setup/master_node')
```

```python
master.head()
```

```python
split = [x.split(',')for x in delay['station']]
```

```python
delay['station_0']=[x[0] for x in split]
delay['station_1']=[x[1] for x in split]
```

```python
del delay['station']
```

```python
master = master[['master_node_id','stop_lat','stop_lon']]
```

```python
master.head()
```

```python
delay
```

```python
delay = delay.replace('Lexington Av-63 St_0','Lexington Av/63 St_0')
delay = delay.replace('5 Av-53 St_0','5 Av/53 St_0')
```

```python
delay = delay.merge(master.rename(columns={'master_node_id':'station_0',
                       'stop_lat':'station_0_lat',
                       'stop_lon':'station_0_lon'}),
            on= 'station_0',how='left')
```

```python
delay = delay.merge(master.rename(columns={'master_node_id':'station_1',
                       'stop_lat':'station_1_lat',
                       'stop_lon':'station_1_lon'}),
           on= 'station_1')
```
